# Survey-Analyse: Einsparoptionen

```python
import matplotlib.pyplot as plt
import file_reader
import config_survey
import model
```

## Daten einlesen

```python
data = file_reader.read_csv_to_dict(config_survey.DATA_PATH)
structure = config_survey.SURVEY_STRUCTURE
print(f"{len(list(data.values())[0])} Antworten eingelesen.")
```

## Daten verarbeiten

```python
questions = []
for q in structure:
    obj = model.QuestionFactory.create_question(q)
    obj.process_data(data)
    obj.calculate_score()
    questions.append(obj)

for obj in questions:
    print(f"Frage: {obj.question_text}")
    print(f"  Ergebnis: {obj.result}")
    print(f"  Scores:   {obj.scores}")
    print()
```

## Visualisierung

```python
total_axes = sum(q.num_axes for q in questions)
fig, axes = plt.subplots(total_axes, 1, figsize=(10, 4 * total_axes))
if total_axes == 1:
    axes = [axes]

idx = 0
for obj in questions:
    n = obj.num_axes
    axes[idx].set_title(obj.question_text, fontsize=11, fontweight='bold')
    obj.visualize(axes[idx:idx + n])
    idx += n

plt.tight_layout()
plt.show()
```
